```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
# Load dataset
df=pd.read_csv('/content/employee_productivity_dataset.csv')
df
```

Q1. Handle Missing Values (Basic Cleaning)

Fill missing values using the following:

Age → Median

Salary → Mean

Hours_Worked_Per_Week → Median

Performance_Score → Mean

Display the dataset after handling missing values.

```python
print("\n" + "=" * 30)
print("Q1. HANDLE MISSING VALUES")
print("=" * 30)

df['Age']= df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(df['Hours_Worked_Per_Week'].median())
df['Performance_Score']=df['Performance_Score'].fillna(df['Performance_Score'].mean())

print("Missing values after filling:")
print(df.isnull().sum())
print(f"\nDataset after handling missing values (first 5 rows):\n{df.head()}")
```

Q2. Label Encoding

Convert these columns into numeric values using Label Encoding:

Gender

Department

Show the updated columns

```python
print("\n" + "=" * 30)
print("Q2. LABEL ENCODING")
print("=" * 30)

le = LabelEncoder()
df['Gender']= le.fit_transform(df['Gender'])
gender_map = dict(zip(le.classes_, le.transform(le.classes_)))
print(f"Gender encoding->{gender_map}")

df['Department'] = le.fit_transform(df['Department'])
dept_map = dict(zip(le.classes_, le.transform(le.classes_)))
print(f"Department encoding -> {dept_map}")

print(f"\nUpdated Gender & Department (first 10 rows):\n{df[['Gender', 'Department']].head(10)}")
```

Q3. One-Hot Encoding
Apply One-Hot Encoding on:
Work_Mode
Location
Display the dataset and check how many new columns are created

```python
print("\n" + "=" * 30)
print("Q3. ONE-HOT ENCODING")
print("=" * 30)

cols_before = df.shape[1]
df = pd.get_dummies(df, columns=['Work_Mode', 'Location'], drop_first=False)
cols_after = df.shape[1]

print(f"Columns before 1 : {cols_before}")
print(f"Columns after 1  : {cols_after}")
print(f"New columns created: {cols_after - cols_before}")
ohe_cols_list = [c for c in df.columns if 'Work_Mode' in c or 'Location_' in c]
print(f"New column names   : {ohe_cols_list}")
print(f"\nFirst 3 rows (OHE columns only):\n{df[ohe_cols_list].head(3)}")
```

Q4. Normalization (Min-Max Scaling)
Normalize the following columns:
Salary
Hours_Worked_Per_Week
Use MinMaxScaler and display results

```python
print("\n" + "=" * 30)
print("Q4. NORMALIZATION (MIN-MAX SCALING)")
print("=" * 30)

scaler_mm = MinMaxScaler()
df[['Salary', 'Hours_Worked_Per_Week']] = scaler_mm.fit_transform(
    df[['Salary', 'Hours_Worked_Per_Week']]
)
print("Salary and Hours_Worked_Per_Week after Min-Max Scaling (range 0–1):")
print(df[['Salary', 'Hours_Worked_Per_Week']].describe().round(4))
print(f"\nFirst 5 rows:\n{df[['Salary', 'Hours_Worked_Per_Week']].head()}")

```

Q5. Standardization (Scaling)
Apply StandardScaler on:
Age
Projects_Completed
Display the transformed values.


```python
print("\n" + "=" * 40)
print("Q5. STANDARDIZATION (STANDARD SCALER)")
print("=" * 40)

scaler_std = StandardScaler()
df[['Age','Projects_Completed']] = scaler_std.fit_transform(
    df[['Age','Projects_Completed']]
)
print("Age and Projects_Completed after StandardScaler (mean≈0, std≈1):")
print(df[['Age', 'Projects_Completed']].describe().round(4))
print(f"\nFirst 5 rows:\n{df[['Age', 'Projects_Completed']].head()}")
```

Q6. Compare Scaling Methods
Apply both:
MinMaxScaler
StandardScaler
On the Salary column.
Show both results side by side

```python
print("\n" + "=" * 60)
print("Q6. COMPARE SCALING METHODS ON SALARY")
print("=" * 60)

df_orig = pd.read_csv('/content/employee_productivity_dataset.csv')
df_orig['Salary'] = df_orig['Salary'].fillna(df_orig['Salary'].mean())

salary_vals = df_orig[['Salary']].copy()
salary_minmax   = MinMaxScaler().fit_transform(salary_vals)
salary_standard = StandardScaler().fit_transform(salary_vals)

comparison = pd.DataFrame({
    'Original_Salary': salary_vals['Salary'].values,
    'MinMax_Scaled'  : salary_minmax.flatten().round(4),
    'Standard_Scaled': salary_standard.flatten().round(4)
})
print(comparison.head(10).to_string(index=False))
print(f"\nSummary Stats:\n{comparison.describe().round(4)}")
```

Q7. Build Preprocessing Pipeline
Create a pipeline that:
Applies encoding to categorical columns
Applies scaling to numerical columns
Use:
ColumnTransformer
Pipeline


Q8. Apply Pipeline
Apply the pipeline on the dataset.
Display:
Transformed dataset
Shape of final dataset

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
#Load Dataset
df=pd.read_csv('/content/employee_productivity_dataset.csv')
print(df)

# Numerical Columns
numerical_cols = [
    'Age',
    'Hours_Worked_Per_Week',
    'Projects_Completed',
    'Performance_Score',
    'Salary'
]

# Categorical Columns
categorical_cols = [
    'Employee_ID',
    'Name',
    'Department',
    'Gender',
    'Work_Mode',
    'Join_Date',
    'Location'
]

# Numerical Pipeline
num_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

# Categorical Pipeline
cat_pipeline = Pipeline([
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# ColumnTransformer
preprocessor = ColumnTransformer([
    ('num', num_pipeline, numerical_cols),
    ('cat', cat_pipeline, categorical_cols)
])

# Apply Pipeline
transformed_data = preprocessor.fit_transform(df)

# Display Result
print(transformed_data)

# Display Shape
print("Shape of Final Dataset:", transformed_data.shape)
```
